```python
import numpy as np
import pandas as pd
import statistics
from sklearn.model_selection import train_test_split

dataset=pd.read_excel('data_uts.xlsx')
A=dataset.iloc[:,1:]
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values

# Splitting Data
X_latih, X_uji, Y_latih, Y_uji = train_test_split(X, Y, test_size=0.2, random_state=1)
print("Seluruh Data : \n", A, "\n")
print("Data X latih : \n", X_latih, "\n")
print("Data X uji : \n", X_uji, "\n")
print("Data Y latih : \n", Y_latih, "\n")
print("Data Y uji : \n", Y_uji, "\n")

# Normalisasi data latih
def normalisasi(w):
  x,y=np.shape(w)
  R=[]
  for i in range(x):
    r=[]
    for j in range(y):
      if(np.max(w[:,j])-np.min(w[:,j])) !=0:
        r1= (2*((w[i,j]-np.min(w[:,j])) / (np.max(w[:,j]) - np.min(w[:,j]))))-1
      else:
        r1=0
      r.append(r1)
    R.append(r)
  R=np.array(R).astype(float)
  return R

# Nilai min data latih
def normalisasi_min(w):
  x,y=np.shape(w)
  R=[]
  for j in range(y):
     r1= np.min(w[:,j])
     R.append(r1)
  R=np.array(R).astype(float)
  return R

# Nilai max data latih
def normalisasi_max(w):
  x,y=np.shape(w)
  R=[]
  for j in range(y):
     r1= np.max(w[:,j])
     R.append(r1)
  R=np.array(R).astype(float)
  return R

# Normalisasi Data Uji
def normalisasi_uji(datauji,min,max):
  x,y=np.shape(datauji)
  r1= (2*(datauji-min)/(max - min))-1
  R=np.array(r1).astype(float)
  return R
min= normalisasi_min(X_latih)
max= normalisasi_max(X_latih)
X_latih= normalisasi(X_latih)
X_uji= normalisasi_uji(X_uji,min,max)
print("Data X latih normalisasi: \n",X_latih, "\n")
print("Data X uji normalisasi: \n",X_uji, "\n")

# Jarak
def jarak(x1,x2):
  return np.sqrt(np.sum((x1-x2)**2))

m1,n1=np.shape(X_uji)
R=[]
for i in range(m1):
  print("=========================================================")
  print("Data uji ke-", i+1)
  x,y=np.shape(X_latih)
  r=[]
  for j in range(x):
    r1=jarak(X_uji[i],X_latih[j])
    r.append(r1)

  result=pd.DataFrame(r)
  result["Status Gizi"]=Y_latih
  print("Perhitungan jarak : \n",result)
  print()
  h_k=result.nsmallest(10,[0])
  print("Ambil k terdekat : \n",h_k)
  modus = statistics.mode(h_k["Status Gizi"])
  print("Sehingga hasilnya adalah: status gizi balita ",modus, "\n")
  R.append(modus)

print("Data Y uji: \n", Y_uji, "\n")
print("Prediksi data Y uji : \n",R, "\n")
prediksiBenar = (R == Y_uji).sum()
prediksiSalah = (R != Y_uji).sum()
print("Prediksi Benar : ", prediksiBenar, " data")
print("Prediksi Salah : ", prediksiSalah, " data")
akurasi = prediksiBenar/(prediksiBenar + prediksiSalah) * 100
print("Akurasi ", akurasi, " %")
```
